```python
data = pd.read_csv("16k_Movies.csv")
data.head()
```

```python
# Remove unecessary column: 'Unnamed: 0'
data.drop(columns = ["Unnamed: 0"], inplace = True)
data.head()
```

What was the highest rated movie per year? 

```python
# Convert dates to pandas datetime timestamps 
data["Release Date"] = pd.to_datetime(data["Release Date"])
data.head()
```

```python
# Create a year column 
data["Year"] = data["Release Date"].dt.year
data.head()
```

```python
# Get a list of indicies that have the highest rated moives 
highest_ratings_idx = data.groupby(["Year"])["Rating"].idxmax()
highest_rated_movies = data.iloc[highest_ratings_idx]
highest_rated_movies.reset_index(drop = True, inplace = True)
highest_rated_movies
```

Which director has released the most number of movies over the years?

```python
# Groupby directors and calculate how many movies they have released using "size" function 
director_count = data.groupby("Directed by").agg(Num_Movies = pd.NamedAgg("Directed by", "size")).sort_values("Num_Movies", ascending = False).reset_index(drop = False)
director_count
```

Does a particular genre always have a high rating? If so, what genre?

```python
# Groupby genre and rating 
# Put results into a new dataframe
highest_rated_genre = pd.DataFrame(data.groupby("Genres")["Rating"].max()).sort_values("Rating", ascending = False)
highest_rated_genre.dropna(inplace = True)
highest_rated_genre
```
